```python
import pandas as pd
from operator import itemgetter
```

```python
daylight_saving_date = ['2016-03-27', 
                        '2017-03-26',
                       '2018-03-25']

daylight_saving_time = ['02', '02', '02']
```

```python
data = pd.read_csv('weather_data.csv')
date = data.date.tolist()
time = data.time.tolist()

process_unique_date =()

unique_date = list(set(date))

for x in unique_date:
    process_unique_date += (x.split('-'),)

process_unique_date = sorted(process_unique_date, key=itemgetter(0, 1, 2))

s = '-'
unique_date = []
for x in process_unique_date:
    temp = (x[0], x[1], x[2])
    unique_date.append(s.join(temp))
    
for x in unique_date:
    if date.count(x) != 48:
        
        output = []
        temp = []
        
        for idx, y in enumerate(date):
            if y == x:
                temp.append(time[idx])
        
        process_hour = ()
        process_minute = ()
        for y in temp:
            process_hour += (y.split(':')[0],)
            process_minute += (y.split(':')[1],)
        
        for y in range(24):
            
            if process_hour.count(str(y).zfill(2)) > 2:
                output.append('This hour has more than 2 items: {} {}'.format(x, str(y).zfill(2)))
            
            if 0 < process_hour.count(str(y).zfill(2)) < 2:
                
                for idx, temp in enumerate(process_hour):
                    if int(temp) == y:
                        if process_minute[idx] == '00':
                            temp_minute = '30'
                        else:
                            temp_minute = '00'
                        
                output.append('Missing value: {} {}:{}:00'.format(x, 
                                                                  str(y).zfill(2), 
                                                                  temp_minute))
            daylight_checker = False
            if process_hour.count(str(y).zfill(2)) == 0:
                for dl_date, dl_time in zip(daylight_saving_date, daylight_saving_time):
                    if x == dl_date and str(y).zfill(2) == dl_time: daylight_checker = True
                if not daylight_checker:
                    output.append('This hour has no item: {} {}'.format(x, str(y).zfill(2)))
                    
        for y in output: print(y)
        if output: print('-' * 80)
```
